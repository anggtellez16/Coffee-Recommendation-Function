```python
import pandas as pd
import numpy as np
```

```python
coffee_ratings = pd.read_csv("coffee_ratings.csv")
coffee_ratings.count() ## no need to use dropna
```

```python
coffee_ratings.describe()
```

```python
features = ["flavor", "aroma", "aftertaste", "acidity", "body"]
items_df = coffee_ratings[features] 
```

```python
items_df ## Cleaned Data
```

```python
random_flavor = np.random.normal(7.520426, 0.398442, size=(5, 1)) 
random_aroma = np.random.normal(7.566706, 0.377560, size=(5, 1)) 
random_aftertaste = np.random.normal(7.401083, 0.404463, size=(5, 1)) 
random_acidity = np.random.normal(7.535706, 0.379827, size=(5, 1)) 
random_body = np.random.normal(7.517498, 0.370064, size=(5, 1)) 
```

```python
user = np.hstack([random_flavor, random_aroma, random_aftertaste, random_acidity, random_body])
user_df = pd.DataFrame(user) ## User - Rating Matrix
```

```python
def recommendation(item_matrix, user_matrix):

    ## Building the Resulting Matrix
    Result = np.empty((len(user_matrix), len(item_matrix)))
    for i in range(0,(len(user_matrix))):
        user_vec = user_matrix.iloc[i]
        user_vec_norm = user_vec/np.linalg.norm(user_vec)
        for j in range(0,(len(item_matrix))):
            item_vec = item_matrix.iloc[j]
            item_vec_norm = item_vec/np.linalg.norm(item_vec)
            dot_product = np.dot(user_vec_norm, item_vec_norm)
            Result[i][j] = float(dot_product)

    Result_df = pd.DataFrame(Result) ## Result Matrix

    ## Columns represent name of items with max row values. 
    max_col = Result_df.idxmax(axis=1) 
    
    ## Building Recommendation Table to display.
    final = pd.DataFrame()
    ind = []
    for i in range(len(max_col)):
        final[i] = item_matrix.iloc[max_col.iloc[i]]
        ind.append(str(max_col.iloc[i]))

    feature = list(item_matrix.columns)
    final.index = ind
    final.columns = list(item_matrix.columns)
    return final
```

```python
recommendation(items_df, user_df)
```
